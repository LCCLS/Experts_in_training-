```python
#Exercise 2) List Methods

#a.) 
#append - takes an item (positional argument) and adds it to a list; goal - returning a list with the added item 
# takes no keyword arguments. 
#working example: 
list = []
variable = 1
list.append(variable)
print(variable)
print(list)
print()

#b.) 
#count - counts how many times an item is repeated in a list. goal - returning an integer 
#representing the count of an item. param - item in the list that is supposed to be counted (positional argument).
#working example:
list = [1,2,3,4,1,2,34,1,12,3,12,12,12,1,2,2,2,1,1,1]
print(list)
print(list.count(1))
print()
#c.) 
#index - returns the index (lowest one) in which an intem appears in. goal - returning an integer representing the
#first index of an intem. param - the item that should be analysed for its index. (positional)
#working example: 
list = [1,2,3,4,5,6,1,2,3,4,5,6]
print(list)
print(list.index(5))
print()
#d.) 
#insert - inserting an intem to a list. first the index is set (positional), and then the item itself(also positional).
# then when the list is returned again, it should be updated (goal).
#working example: 
list = [1,2,3,4,5,6]
print(list)
list.insert(0,0)
print(list)
print()
#e.) 
#pop - pops off the last item of a list. it takes the positinal parameter of an item but it is not necessary to do it. 
# the goal is to return a list with the last item missing. 
#working example: 
list = [1,2,3,4,5]
print(list)
list.pop()
print(list)
```
